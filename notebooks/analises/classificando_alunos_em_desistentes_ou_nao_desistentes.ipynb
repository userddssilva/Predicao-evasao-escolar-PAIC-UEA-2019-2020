## Ordenando as avaliações por data de inicio e retirando o laboratório 0, desafios e avaliações do tipo homework

### Tirando todos desafios, avaliação 01 e avaliação 1

````
def drop_homeworks(homework_name):
    if not re.search(r'desafio|avaliação\s\d+', homework_name, re.IGNORECASE):
        return True
    else:
        return False

# drop_homeworks('Desafios 7: Matrizes')
````

### Removendo o laboratório 0 das turmas

````
def lab_zero(row_title):
    find_zero = re.search(r'\s0\s|lab0 ', row_title, re.IGNORECASE)
    if find_zero is not None:
        return False
    else:
        return True
````

## Montando um dataset de alunos com notas das avaliações

### Lendo os arquivos e tirando os dados de cada avaliação

````
def read_grade(path_grade, grade_file):
    grade_path = os.path.join(path_grade, grade_file)
    grade = None
    numb_exercise = None
    correct = None
    incorrect = None
    blank = None
    with open(grade_path, "r") as fl:
        for line in fl.readlines():
            if grade is None: grade = re.search("---- grade \(\d+\-\d+\)\: (.+)", line)
            if numb_exercise is None: numb_exercise = re.search("---- number of exercises: (.+)", line)
            if correct is None: correct = re.search("---- correct: (.+)", line)
            if incorrect is None: incorrect = re.search("---- incorrect: (.+)", line)
            if blank is None: blank = re.search("--- blank: (.+)", line)
        
            if grade is not None and not isinstance(grade, str): grade = grade.group(1)
            if numb_exercise is not None and not isinstance(numb_exercise, str): numb_exercise = numb_exercise.group(1)
            if correct is not None and not isinstance(correct, str): correct = correct.group(1)
            if incorrect is not None and not isinstance(incorrect, str): incorrect = incorrect.group(1)
            if blank is not None and not isinstance(blank, str): blank = blank.group(1)
    return grade, numb_exercise, correct, incorrect, blank

read_grade('codebench/2019-1/230/users/4317/grades/', '1576.log')
````

### Consultando os dados das avaliações

````
def consult_type(assessment):
    value = df_assessments[df_assessments['assessment'] == int(assessment)]
    value = value["assessment_type"].values[0]
    return value
consult_type(380)
````

````
def consult_title(assessment):
    line = df_assessments[df_assessments['assessment'] == int(assessment)]
    value = line['title'].values[0]
    return value
consult_title(380)
````

````
def amount_homework(class_id):
    return df_assessments[(df_assessments['class'] == int(class_id))  & \
                          (df_assessments['assessment_type'] == 'homework')].count().values[0]
````

````
def amount_exam(class_id):
    return df_assessments[(df_assessments['class'] == int(class_id))  & \
                          (df_assessments['assessment_type'] == 'exam')].count().values[0]
````

### Extraíndo os dados dos alunos em relação as suas avaliações

````
import re
import os

root = 'codebench/'
features = ['semester_id', 
            'class_id', 
            'user_id', 
            'assessment_id', 
            'grade',
            'amount_exercise',
            'number_of_correct_answers',
            'number_of_incorrects_answers', 
            'number_of_blank_answers',
            'assessment_type',
            'assessment_title',
            'amount_homework',
            'amount_exam'
            ]

df_user_grade = []


cont_p = 0
cont_n = 0

# percorrendo os semestres
for semester_id in os.listdir(root):
    path_class = os.path.join(root, semester_id)

    #percorrendo todas turmas
    for class_id in os.listdir(path_class):
        path_user = os.path.join(path_class, class_id, 'users/')
        
        # quantidade de avaliações do tipo exam
        amount_exam_per_class = amount_exam(class_id)

        #quantidade de avaliações do tipo homework
        amount_homework_per_class = amount_homework(class_id)

        # percorrendo todos os alunos
        for user_id in os.listdir(path_user):
            path_grade = os.path.join(path_user, user_id, 'grades/')
            
            
            # verificando se o aluno fez alguma avaliação
            if len(os.listdir(path_grade)) > 0:
                cont_p += 1
                for assessment_id in os.listdir(path_grade):
                    if 'final' not in assessment_id:
                        
                        # dados das avaliações feitas pelo aluno
                        grade, numb_exercise, correct, incorrect, blank = read_grade(path_grade, assessment_id)
                        assessment_id = assessment_id.replace('.log', '')
                        assess_type = consult_type(assessment_id)
                        assess_title = consult_title(assessment_id)

                        # montando a estrutura com os dados
                        values = [semester_id, class_id, user_id, 
                                    assessment_id, grade, numb_exercise, correct,
                                    incorrect, blank, assess_type, assess_title, 
                                    amount_homework_per_class, amount_exam_per_class]
                        row = dict(zip(features, values))
                        df_user_grade.append(row)
            else:
                cont_n += 1
                values = [semester_id, class_id, user_id, '', '', '', '', '', '', '', '',
                           amount_homework_per_class, amount_exam_per_class]
                row = dict(zip(features, values))
                df_user_grade.append(row)

df_user_grade = pd.DataFrame(df_user_grade)
print(cont_p, cont_n, cont_p+cont_n)
````

````
df_all_users = df_user_grade[['semester_id','class_id','user_id', 'amount_exam']].groupby(by=['semester_id','class_id','user_id'])['amount_exam'].count().reset_index()
df_all_users.drop('amount_exam', axis=1, inplace=True)

df_user_with_exam = df_user_grade[['semester_id','class_id','user_id', 'assessment_type']]
df_user_with_exam = df_user_with_exam[df_user_with_exam['assessment_type'] == 'exam']
df_user_with_exam = df_user_with_exam.groupby(by=['semester_id','class_id','user_id'])['assessment_type'].count().reset_index()
df_user_with_exam.rename({'assessment_type':'amount_exam'}, axis=1, inplace=True)
# df_user_with_exam.drop('assessment_type', axis=1, inplace=True)
````

## Verificando se o aluno fez 80% dos exercícios

````
def zero_perct(group):
    amount_numbers = group.grade.count()
    count_zero = 0
    for g in group.grade:
        if float(g) == 0.0:
            count_zero += 1
    if amount_numbers * 4 <= count_zero * 5:
        return 0
    else:
        return 1
df_zero_perct = df.groupby(['semester_id', 'class_id', 'user_id']).apply(zero_perct).to_frame().reset_index()
````

````
df_zero_perct.rename({0:'less_than_80_percent'}, axis=1, inplace=True)
df_zero_perct
````

## Agrupando os dois critérios

````
df_classes = pd.merge(df_amount_exam_by_user, df_zero_perct)
df_classes
````

## Classificando os registros

````
def classification(pr_1, pr_2):
    if pr_1 == 1 and pr_2 == 1:
        return 'Não Desistente'
    else:
        return 'Desistente'

df_classes['target'] = df_classes.apply(lambda row: classification(row['did_more_than_2'], row['less_than_80_percent']), axis=1)
df_classes
````
