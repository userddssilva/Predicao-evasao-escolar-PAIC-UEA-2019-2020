# Selecionando as turmas de semestre impar

## Importando o dataset de avaliações válida

```
import re
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
```

```
df_valid_assessment = pd.read_csv('/content/drive/Shared drives/iniciacao_cientifica_2019_2020/csv_datasets/avaliacoes_validas.csv', sep=';')
df_valid_assessment.rename({'semester':'semester_id', 'class':'class_id', 'assessment':'assessment_id'}, axis=1, inplace=True)
df_valid_assessment = df_valid_assessment.iloc[: , [0, 1, 2, 7, 3, 4]]
```

## Criando uma função de filtro para capturar as turmas ímpares

```
def select_impar_semester(semester):
    if (re.search(r'\-1', semester)) :
        return True
    else:
        return False

fltr = df_valid_assessment['semester_id'].apply(select_impar_semester)
df_assessment_first_semester = df_valid_assessment[fltr]
```

```
df_assessment_first_semester
```

## Visualizando os semesters

```
df_assessment_first_semester['semester_id'].value_counts()
```

## Visualizando a quantidade de assessments válidos

```
df = df_assessment_first_semester.groupby(by=['semester_id', 'class_id'])['assessment_id'].count().to_frame().reset_index()
df.rename({'assessment_id':'amount_assessment'}, axis=1, inplace=True)
df
```

## Assessments do tipo **homework**

```
df_homework = df_assessment_first_semester.loc[df_assessment_first_semester['assessment_type'] == 'homework']
df_homework
```

## Quantidade de assessment do tipo **homework** por turma

```
df = df_homework .groupby(by=['semester_id', 'class_id'])['assessment_id'].count().to_frame().reset_index()
df.rename({'assessment_id':'amount_assessment'}, axis=1, inplace=True)
df
```
