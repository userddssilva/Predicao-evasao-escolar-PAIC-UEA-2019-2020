## Criação dos scripts

Cada Script foi gerado com base em dois parâmetros principais: A Data de inicio e fim. Dentro dessa janela de tempo foram extraídas as **features** para cada usuário de cada turma.

- As datas de inicio e fim correspondem as datas iniciais da primeira avaliação válida que pode ser do tipo **exam** ou **homework** até o final de uma avaliação **X**.

- Cada turma possui uma quantidade de avaliações diferentes, algumas tem 4 outras possuem 10 e outras váriam dentro desse intervalo.

## Lista de Features extraída


**semester_id:** Semestre que o aluno está matriculado.

**class_id:** Número da turma do aluno.

**user_id:** Número de identificação do aluno.

**number_logins:** Quantidade de logins que o aluno realizou dentro do intervalo de tempo. 

**average_active_time:** Média de tempo ativo de tempo dentro da IDE da plataforma codebench (cada interação possui uma data e hora, caso essa data e hora seja maior que 3 min esse intervalo de tempo entre interações é descartado, caso seja maior é acumulado para extração da média)

**average_exam_sub_comp:** Média da quantidade de submissões certas de todas as questões das avaliações do tipo **exam** dentro do intervalo de tempo.

**average_exam_sub_inc:** Média da quantidade de submissões que compilaram, mas não produziram resposta correta de todas as avaliações do tipo **exam** dentro do intervalo de tempo por questão.

**average_exam_sub_err:** Média da quantidade de submissões que produziram algum tipo de erro na compilação de todas as avaliações do tipo **exam** dentro do intervalo de tempo por questão.

**average_exam_sub_mean_time:** Para extrair essa feature, primeiro acumula-se os intervalos de tempo de cada questão das avaliações do tipo **exam**, (cada submissão possui data e hora, acumula-se o intervalo entre essas submissões) depois de acumular extrai-se a média. E então acumula-se as médias de cada questão para em seguida extrair-se a média das médias acumuladas.

**average_exam_test:** Média da quantidade de testes de cada questão das avalições do tipo **exam**.

**average_exam_test_err:** Média da quantidade de testes que produziram algum erro dentro da IDE, para cada questão das avaliações do tipo **exam**.

**average_exam_test_mean_time:** Média das médias acumuladas de cada intervalo de tempo entre os testes das questões das avaliações do tipo **exam**.

**average_exam_code_line:** Média da quantidade de linhas válidas de código escrito de cada questão das avaliações do tipo **exam**. (foram consideradas de linhas válidas, linhas diferentes de comentários e linhas em branco no código).

**average_homework_sub_comp:** Média da quantidade de submissões certas de todas as questões das avaliações do tipo **homework** dentro do intervalo de tempo.

**average_homework_sub_inc:** Média da quantidade de submissões que compilaram, mas não produziram resposta correta de todas as avaliações do tipo **homework** dentro do intervalo de tempo por questão.

**average_homework_sub_err:** Média da quantidade de submissões que compilaram, mas não produziram resposta correta de todas as avaliações do tipo **homework** dentro do intervalo de tempo por questão.

E então acumula-se as médias de cada questão para em seguida extrair-se a média das médias acumuladas.

**average_homework_test:** Média da quantidade de testes que produziram algum erro dentro da IDE, para cada questão das avaliações do tipo **homework**.

**average_homework_test_err:** Média da quantidade de testes que produziram algum erro dentro da IDE, para cada questão das avaliações do tipo **homework**.

**average_homework_test_mean_time:** Média das médias acumuladas de cada intervalo de tempo entre os testes das questões das avaliações do tipo **homework**.

**average_homework_code_line:** Média da quantidade de linhas válidas de código escrito de cada questão das avaliações do tipo **homework**. (foram consideradas de linhas válidas, linhas diferentes de comentários e linhas em branco no código).

**sum_exam_sub_comp:** Soma acumulada de todas submissões que produziram uma resposta certa de cada questão das avaliações do tipo **exam**.

**sum_exam_sub_inc:** Soma acumulada de todas as submissões que produziram uma resposta incorreta de cada questão das avaliações do tipo **exam**.

**sum_exam_sub_mean_time:** soma acumulada das médias dos intervalos de tempo entre cada submissão de cada questão das avaliações do tipo **exam**.
